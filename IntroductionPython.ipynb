```python
identifiers = ['A', '_B', 'C', 'z']  # List of identifiers

keywords = ['def', 'del', 'and', 'for', 'while', 'is', 'raise', 'break', 'else'
          'global', 'not', 'try', 'continue', 'exec', 'import', 'pass'] # list of keywords

operators = ['+', '-', '*', '/', '%', '**', '//', '<<', '>>']

delimiters = ['(', ')', '[', ']', '{', '}']


```

These are all literals in python.
<pre>
42                       # Integer literal
3.14                     # Floating-point literal
1.0j                     # Imaginary literal
'hello'                  # String literal
"world"                  # Another string literal
"""Good
night"""                 # Triple-quoted string literal
</pre>

You can combine literals and delimiters to create more complex structures.

<pre>
[ 42, 3.14, 'hello' ]    # List
( 100, 200, 300 )        # Tuple
{ 'x':42, 'y':3.14 }     # Dictionary
</pre>
