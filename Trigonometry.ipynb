Consider the following triangle: 



What we can see here is that it is a right-triangle given the angle at C. This exercise asks us to find $\cos(\angle BAC + 30^{\circ})$

First we will start by understanding 

* $(\sin(\angle BAC) = \dfrac{\text{opposite}}{\text{hypotenuse}})$
* $(\cos(\angle BAC) = \dfrac{\text{adjacent}}{\text{hypotenuse}})$

Then we need to add 30 degrees using the trig sum identity

$cos(\alpha + \beta) = cos \alpha cos \beta - sin \alpha sin \beta$

That means we get 

$\cos(\angle BAC + 30^{\circ}) = \cos(\angle BAC)\cos 30^{\circ} - \sin(\angle BAC)\sin 30^{\circ}$



```python
# finding the cosine and sine values 

import math

hypotenuse = 20
leg1 = 12
leg2 = 16

opposite = 16
adjacent = 12
sine = opposite/hypotenuse 
cosine = adjacent/hypotenuse 

# finding the special-angle trig values

cos30 = math.cos(math.radians(30))
sin30 = math.sin(math.radians(30))

# calculating the value
cos_res = cosine*cos30 - sine*sin30
```
